## Transfer Learning

Neste trabalho, vamos trabalhar com o conjunto de dados de imagens CIFAR-10. O dataset contém 60000 imagens 32x32 com 3 canais de cores (RGB), sendo 10000 delas um conjunto de teste. Cada imagem pertence a uma de 10 classes, numeradas de 0 a 9.

Para classificação das imagens, usaremos redes neurais convolucionais e transfer learning. Para visualização dos resultados e treinamento das redes, vamos plotar gráficos mostrando as acurácias nos conjuntos de treino e teste em cada época do treinamento. Para comodidade, vamos definir uma função que realiza essa plotagem, a partir de um objeto `History` do `Keras`.

```python
def plot_history(history, title):
  # Acurácia
  plt.plot(history.history['acc'])
  plt.plot(history.history['val_acc'])
  plt.title(title + ': Acurácia')
  plt.ylabel('Acurácia')
  plt.xlabel('Época')
  plt.xticks(np.arange(0,21,2))
  plt.legend(['Treinamento', 'Teste'], loc='upper left')
  plt.show()
  
  # Função de perda
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title(title+ ': Perda')
  plt.ylabel('Perda')
  plt.xlabel('Época')
  plt.xticks(np.arange(0,21,2))
  plt.legend(['Treinamento', 'Teste'], loc='upper left')
  plt.show();
```

Agora vamos carregar o dataset, presente no módulo `keras.datasets`.

```python
from keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

_A primeira etapa do trabalho consiste em dividir o conjunto de dados em duas partes A e B. Na parte B você irá selecionar todas as imagens pertencentes a 2 das 10 classes e na parte A ficarão as imagens das demais 8 classes. As classes selecionadas na parte B serão as de número x e y, onde x e y são os dois últimos dígitos do seu número de matrícula._

Vamos separar os dados em A e B. Na verdade, os vetores `A/B_train/test` conterão os índices, nos vetores `X/y_train/test`, dos elementos que devem pertencer a cada conjunto.

```python
B_train = np.array([idx for idx, y in enumerate(y_train) if y[0] in (4,2)]) # 2017014642
A_train = np.array([idx for idx, y in enumerate(y_train) if idx not in B_train])

B_test = np.array([idx for idx, y in enumerate(y_test) if y[0] in (4,2)]) # 2017014642
A_test = np.array([idx for idx, y in enumerate(y_test) if idx not in B_test])
```

_Feito isto, você deverá treinar uma CNN para classificação das imagens usando os dados da parte A e a API do Keras para o TensorFlow. A CNN deverá ter quatro camadas de convolução e duas camadas de Max Pooling (não necessariamente nessa ordem) e duas camadas Fully Connected ao final._

Primeiro, devemos converter a representação das classes nos vetores `y` para o formato One-Hot, para alimentar nossas redes.

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
```

---

### CNN para os dados da parte A

Definimos uma função para criação do modelo com alguns parâmetros, para conveniência.

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Dense, Dropout, Activation, Flatten

def model_A(img_shape, num_classes,
                 dropout_conv=0.25, dropout_classif=0.25):
    model = Sequential()
    
    # Convolução
    model.add(Conv2D(filters=16, 
                     kernel_size=(3, 3),
                     padding='same',
                     input_shape=img_shape,
                     activation='relu'))
    
    # Convolução
    model.add(Conv2D(filters=32,
                     kernel_size=(3, 3),
                     padding='same',
                     activation='relu'))
    
    # Max Pooling
    model.add(MaxPooling2D(pool_size=2))
    model.add(Dropout(dropout_conv))
  
    # Convolução
    model.add(Conv2D(filters=64,
                     kernel_size=(3, 3),
                     padding='same',
                     activation='relu'))
    
    # Convolução
    model.add(Conv2D(filters=128,
                     kernel_size=(3, 3),
                     padding='same',
                     activation='relu'))
    
    # Max Pooling
    model.add(MaxPooling2D(pool_size=2))
    model.add(Dropout(dropout_conv))
  
    # Fully Connected
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(dropout_classif))
    model.add(Dense(num_classes, activation='softmax'))
    
    return model
```

```python
from keras.optimizers import rmsprop

modelA = model_A(x_train.shape[1:], 8)
opt = rmsprop(lr=1e-5, decay=1e-6)
modelA.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
history_A = modelA.fit(x_train[A_train], y_train[A_train][:,[0,1,3,5,6,7,8,9]], epochs=20, validation_data=(x_test[A_test], y_test[A_test][:,[0,1,3,5,6,7,8,9]]), verbose=0)
plot_history(history_A, "Rede com os dados A")
```

Podemos ver que a função de perda, para ambos os conjuntos, estão convergindo. E como a acurácia no conjunto de teste está até melhor do que no conjunto de treinamento, o modelo está regularizado (embora pareça que à medida que aumentamos o número de épocas, haja overfitting).

Os números de filtros para as camadas de convolução foram escolhidos com a intenção de criar feature maps cada vez mais específicos.

---

### Sem Transfer Learning:  Treinar rede para os dados da parte B do zero, inicializando os pesos aleatoriamente.

Por se tratar de um problema menos complexo, a criação de feature maps mais simples por um número reduzido de filtros se mostrou mais apropriada.

```python
from keras.optimizers import sgd

def model_B1(img_shape, num_classes,
                 dropout_conv=0.25, dropout_classif=0.25):
    model = Sequential()
    
    # Convolução
    model.add(Conv2D(filters=32, 
                     kernel_size=(3, 3),
                     padding='same',
                     input_shape=img_shape,
                     activation='relu'))
    
    # Convolução
    model.add(Conv2D(filters=32,
                     kernel_size=(3, 3),
                     activation='relu'))
    
    # Max Pooling
    model.add(MaxPooling2D(pool_size=2))
    model.add(Dropout(dropout_conv))
  
    # Convolução
    model.add(Conv2D(filters=64,
                     kernel_size=(3, 3),
                     padding='same',
                     activation='relu'))
    
    # Convolução
    model.add(Conv2D(filters=64,
                     kernel_size=(3, 3),
                     activation='relu'))
    
    # Max Pooling
    model.add(MaxPooling2D(pool_size=2))
    model.add(Dropout(dropout_conv))
  
    # Fully Connected
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(dropout_classif))
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

modelB1 = model_B1(x_train.shape[1:], 2, dropout_conv=0.25, dropout_classif=0.5)
opt = rmsprop(lr=5e-5, decay=1e-6)
modelB1.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
history_B1 = modelB1.fit(x_train[B_train], y_train[B_train][:,[2,4]], epochs=20, validation_data=(x_test[B_test], y_test[B_test][:,[2,4]]), verbose=0)
plot_history(history_B1, "Rede treinada do zero com os dados B")
```

Para essa rede, foi mais difícil explorar uma configuração que convergisse e fosse mais regularizada. Mas por se tratar de um problema binário e não multinomial, a rede atingiu acurácia superior à treinada para os dados da parte A.

---

### Fine-tuning em uma camada: Utilizar os pesos da  rede  treinada na parte A e realizar fine-tuning na última camada Fully Connected.

O transfer learning consiste em reaproveitar os pesos de redes previamente treinadas para conjuntos de dados similares. Embora para os dados da parte B tenhamos duas possíveis classes que não foram consideradas no treinamento da primeira rede, as camadas de convolução e max pooling consistem na extração de features das imagens. Estas mesmas features serão extraídas das imagens do conjunto B, e com fine-tuning na última camada, ajustaremos apenas os pesos da camada softmax, para ativação das classes 4 e 2.

```python
from keras.models import clone_model

modelB2 = Sequential()
for layer in modelA.layers[:-1]: # Fine tuning na última camada fully connected
  modelB2.add(layer)
for layer in modelB2.layers:
  layer.trainable = False
modelB2.add(Dense(2, activation='softmax'))

opt = rmsprop(lr=5e-5, decay=1e-5)
modelB2.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
history_B2 = modelB2.fit(x_train[B_train], y_train[B_train][:,[2,4]], epochs=20, validation_data=(x_test[B_test], y_test[B_test][:,[2,4]]), verbose=0)
plot_history(history_B2, "Rede construída por transfer learning com fine-tuning na última camada")
```

Percebe-se que a acurácia obtida é mais semelhante àquela para os dados da rede A. Pelo transfer learning, estamos limitados à qualidade prévia da extração das features.

Também por estarmos treinando a camada de ativação com as camadas de convolução fixas após treinamento de 20 épocas com os dados de A, a acurácia para as classes de B é maior que a da primeira, para as classes de A. Os feature maps já estão melhor definidos que no começo do treinamento.

---

### Fine-tuning em duas camadas: Utilizar os pesos da rede treinada na parte A e realizar fine-tuning nas duas camadas Fully Connected.

Para essa rede, treinaremos também a outra camada densa fully connected, além da última, para os dados de B. Esperamos atingir maior acurácia, uma vez que os pesos dessa outra camada serão especializados na interpretação de features para identificação das classes 4 e 2, não presentes em seu treinamento original.

```python
modelB3 = Sequential()
for layer in modelA.layers[:-3]: # Excluir ultimas duas fully connected e dropout
  modelB3.add(layer)
for layer in modelB3.layers:
  layer.trainable = False
modelB3.add(Dense(256, activation='relu'))
modelB3.add(Dropout(0.5))
modelB3.add(Dense(2, activation='softmax'))

opt = rmsprop(lr=1e-5, decay=1e-6)
modelB3.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
history_B3 = modelB3.fit(x_train[B_train], y_train[B_train][:,[2,4]], epochs=20, validation_data=(x_test[B_test], y_test[B_test][:,[2,4]]), verbose=0)
plot_history(history_B3, "Rede construída por transfer learning com fine-tuning nas duas últimas camadas FC")
```

Como esperado, atingimos maior acurácia que na outra rede com transfer learning. Porém, exploração de configurações para essa rede foi mais difícil, e pelos resultados, a regularização atingida foi bem pior se comparada às anteriores. Em comparação à primeira rede, tentei diminuir o número de neurônios na penúltima camada fully connected a fim de evitar overfitting, já que se trata de um problema mais simples.

## Conclusão

O melhor resultado obtido na tarefa classificação dos dados B foi usando uma rede treinada do zero e construída para o problema. Esse resultado faz sentido, uma vez que teremos todos pesos ajustados especificamente para os erros na classificação de imagens 4 e 2.

O transfer learning se mostrou uma estratégia válida para reaproveitamento de redes convolucionais previamente treinadas. Embora não tenhamos problemas com o treinamento de uma rede para B do zero, o ajuste de CNNs pode levar **muito** tempo e requerer **grandes** volumes de dados.
